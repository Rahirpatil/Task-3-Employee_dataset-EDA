```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds= pd.read_csv("D:\\Internship\\Machine learning\\employee_dataset_for_eda-task3.xlsx.csv")
print(ds)
```

```python
sns.countplot(data=ds,x="Department",palette='bright')
plt.title("Count of employee as per the department")
```

There are more emoployees in finance department.

```python
sns.countplot(data=ds,x="Gender",palette="Set2")
plt.title("Gender distribution")
```

There are more count of females more than males

```python
sns.countplot(data=ds,x="Location",palette="muted")
plt.title("Location count")
plt.xticks(rotation=45)
```

seattle have more count the of employees

```python
sns.histplot(ds['Age'],bins=10,kde=True)
plt.title('Age Distribution')
```

highest age of employee in range 45-50 followed by the range 25-30.

```python
sns.histplot(data=ds,x="Years_of_Experience",bins=15,kde=True)
plt.title("Years of experience")
```

```python
sns.barplot(data=ds,x="Department",y="Monthly_Salary",palette='viridis')
plt.title("average salary by departments")
```
